```python
# Dependencies
import pandas as pd
from pathlib import Path
# Create a reference the CSV file desired
csv_path = Path("speciesIL.csv")

# Read the CSV into a Pandas DataFrame
illinois_df = pd.read_csv(csv_path, low_memory=False)

# Print the first five rows of data to the screen
illinois_df.head()
# Rename mistyped columns "Aid Given or Received Code " and "Propery Loss"
illinois_df = illinois_df.rename(columns={"Scientific Name": "scientific_name", 
                                    "Common Name": "common_name",
                                    "Where Listed": "where_listed",
                                    "Region  ": "region",
                                    "ESA Listing Status  ": "ESA_listing_status",
                                    "Group": "group_name"})
illinois_df.head()
# export to replace current CSV
illinois_df.to_csv('speciesIL.csv', encoding='utf-8', index=False, header=True)
```
